```python
from geopy.distance import geodesic
# Function to calculate distance using Haversine formula
def calculate_distance(site_lat, site_lon, proximity_lat, proximity_lon):
    site_coords = (site_lat, site_lon)
    proximity_coords = (proximity_lat, proximity_lon)
    distance_km = geodesic(site_coords, proximity_coords).kilometers
    return distance_km

# Example: Calculate distance between Launch Site 1 and Proximity 1
launch_site_lat = spacex_df.loc[0, 'Lat']
launch_site_lon = spacex_df.loc[0, 'Long']

proximity_lat = spacex_df.loc[0, 'Lat']
proximity_lon = spacex_df.loc[0, 'Long']

distance_km = calculate_distance(launch_site_lat, launch_site_lon, proximity_lat, proximity_lon)

print(f"Distance between Launch Site 1 and Proximity 1: {distance_km:.2f} km")
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
